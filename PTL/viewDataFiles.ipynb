```python
# Import necessary libraries
import pydarnio
import os
import bz2
import numpy
# Define the path to the sample data folder
sample_data_folder = 'sampleData'

# List all files in the sample data folder
data_files = [f for f in os.listdir(sample_data_folder) if os.path.isfile(os.path.join(sample_data_folder, f))]
def print_records(records):
    for record in records:
        # print(f"Record: {record.keys()}")
        
        if record["start.year"] == 2009 and record["start.month"] == 1 and record["start.day"] == 3 and record["start.hour"] == 1 and record["start.minute"] == 20:
            if record["end.year"] == 2009 and record["end.month"] == 1 and record["end.day"] == 3 and record["end.hour"] == 1 and record["end.minute"] == 22:
            
                for key in record.keys():
                    if record[key] is not None:
                        if isinstance(record[key], numpy.int16):
                            print(f"{key}: {record[key]}")
                        elif isinstance(record[key], numpy.ndarray):
                            print(f"{key}: {record[key].shape}")
                        elif isinstance(record[key], dict):
                            print(f"{key}: {record[key].keys()}")
                        elif isinstance(record[key], str):
                            print(f"{key}: {record[key]}")
                        elif isinstance(record[key], int):
                            print(f"{key}: {record[key]}")
                        elif isinstance(record[key], float):
                            print(f"{key}: {record[key]}")
                        else:
                            print(f"{key} is type  {type(record[key])}")
        print("\n")
# Function to read and display data from a file
def display_rawacf_data(file_name):
    file_path = os.path.join(sample_data_folder, file_name)
    reader = pydarnio.SDarnRead(file_path)
    records = reader.read_rawacf()
    # print(reader.__dir__())
    print_records(records)

# Function to read and display data from a file
def display_fitacf_data(file_name):
    file_path = os.path.join(sample_data_folder, file_name)
    reader = pydarnio.SDarnRead(file_path)
    records = reader.read_fitacf()
    # print(reader.__dir__())
    print_records(records)

def display_grdmap_data(file_name):
    file_path = os.path.join(sample_data_folder, file_name)
    reader = pydarnio.SDarnRead(file_path)
    records = reader.read_grid()
    #read_map also candidate?
    # print(reader.__dir__())
    print_records(records)
# Display data from each file in the sample data folder
for file in data_files:
    if file.endswith(".bz2"):
        # unzip the file with the bz2 library
        with open(os.path.join(sample_data_folder, file), 'rb') as f:
            data = bz2.decompress(f.read())
            with open(os.path.join(sample_data_folder, file[:-4]), 'wb') as f:
                f.write(data)
        file = file[:-4]
            
    #if file.endswith("rawacf"):
    #    display_rawacf_data(file)
    #if file.endswith("fitacf"):
    #    display_fitacf_data(file)
    if file.endswith("grdmap"):
        print(f"Displaying data from file: {file}")

        display_grdmap_data(file)
```
